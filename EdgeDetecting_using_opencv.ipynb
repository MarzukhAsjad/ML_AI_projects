```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
```

```python
# function to display an image
def display(ImgPath):
    img = cv2.imread(ImgPath)
    cv2_imshow(img)
```

```python
# the edge detect function using cv2 library
def EdgeDetectcv2(ImgPath):
    # read the image
    img = cv2.imread(ImgPath)
    # convert the image to grayscale
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    # convolving the sobel kernel across the image
    # in x direction
    Gx = cv2.Sobel(gray, cv2.CV_64F, 1, 0, ksize=3)
    # in y direction
    Gy = cv2.Sobel(gray, cv2.CV_64F, 0, 1, ksize=3)

    # preprocessing such as absolute valuing the Gx and Gy
    abs_Gx = np.absolute(Gx)
    abs_Gy = np.absolute(Gy)

    # Then we calculate the gradient magnitude
    grad = np.sqrt(np.square(abs_Gx) + np.square(abs_Gy))

    # Then we normalize it
    scaled_grad = np.uint8(255*grad/np.max(grad))

    # We pass it through a threshold to distinguish between weak and strong edges
    weak, strong = cv2.threshold(scaled_grad, 0, 255, cv2.THRESH_BINARY+cv2.THRESH_OTSU)

    # show the image
    cv2_imshow(strong)
```
